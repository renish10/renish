```python
# 11. Identify and remove outliers in the km_driven column using the IQR method.?
# ans. 
df
```

```python
c = df['km_driven']
q1 = c.quantile(0.25)
q2 = c.quantile(0.75)
d = q2-q1
l_bound = q1-1.5*d
u_bound = q2+1.5*d
outliers = df[(df['km_driven']<l_bound) | (df['km_driven']>u_bound)]
d = df[(df['km_driven']>=l_bound) | (df['km_driven']<=u_bound)]
print('data before removing outliers',outliers)
print('data after removing oultiers',d)
```

```python
# 3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
# ans. 
df2
```

```python
data = df2[['Transmission','Price ($)']].dropna()
def cal_iqr(group):
    q1 = np.percentile(group,25)
    q3 = np.percentile(group,75)
    iqr = q3-q1
    return iqr
iqr_val = data.groupby('Transmission')['Price ($)'].apply(cal_iqr)
print('interquartile range for each transmission')
print(iqr_val)
plt.figure(figsize = (10,5))
sns.boxplot(x = 'Transmission',y = 'Price ($)',data = data,palette = 'Set3')
plt.title('price distribution by transmission type')
plt.xlabel('Transmission')
plt.ylabel('price')
plt.show()
```

```python
# insight :- there is a minor difference between auto and manual, as manual values are slighlty up than auto
```

```python
# 7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.?
# ans. 
df4
```

```python
top_5_pop = df4.groupby('Artist')['Popularity'].sum().reset_index()
```

```python
top = top_5_pop.sort_values('Popularity',ascending = False).head(5)
```

```python
top
```

```python
top_tracks = []
for artist in top['Artist']:
    a = df4[df4['Artist'] == artist]
    b = a.sort_values('Popularity',ascending = False).iloc[0]
    top_tracks.append({'Artist' : artist,'track' : b['Track Name']})
top_tracks_df = pd.DataFrame(top_tracks)
print(top_tracks_df)
```
